```python
def apagar(lista):
  x=0
  while(x<len(lista)):
      os.remove(lista[x])
      x=x+1
      
def convert(seconds):
  hrs=strftime("%H:%M:%S.2", gmtime(seconds))
  return hrs

def salvar_arquivo(self, text, filename, name=None):
  self.proxy.salvar_arquivo(text, filename, name)
  lista_mp3.append(filename)

def nova_imagem(nome_arquivo,nome_saida,):
  #comand = 'wkhtmltopdf'
  #url_html=nome_arquivo
  #nome_final=nome_saida
  #opt='--page-width 1080 --page-height 720 --disable-smart-shrinking --encoding UTF-8'
  #'!cat loren.html | wkhtmltoimage --width 1080 --height 720 - test4.jpg'
  #call('cat ',url_html,' | ',comand,'--width 1080 --height 720 - ', nome_final, shell = True)
  cmd = 'cat '+nome_arquivo+' | wkhtmltoimage --width 1080 --height 720 --encoding ISO-8859-1 - '+nome_saida
  #x='cat novo.html | wkhtmltoimage --width 1080 --height 720 --encoding UTF-8 - test6.jpg'
  os.popen(cmd)
  time.sleep(10) 
  #os.system("$comand &url_html $nome_final")
  #!$comand &url_html $nome_final
  #imgkit.from_file(nome_arquivo,nome_saida,config=configs,options=options) #converte o html para imagens
  #lista_jpg.append(nome_saida)

def arrumar_texto(nome_arquivo):
  x1=open(nome_arquivo,mode='r', encoding='utf-8')
  data=x1.read()
  text = re.sub('<[^>]+?>','', data) #limpa as tags de uma string
  return text

def novo_audio(texto_arrumado,nome_sem_formato):
  tts = gTTS(texto_arrumado,lang='pt-br')
  tts.save(nome_sem_formato+'.mp3')
  sound_file = nome_sem_formato+'.mp3'
  time.sleep(10)   
  Audio(sound_file, autoplay=True) 

def tamanho_audio(url):
  audiofile = AudioFileClip(url)
  tempo = (audiofile.duration)
  print(tempo)
  return tempo

def novo_frame(frame,audio,nome):
  time=tamanho_audio(audio)
  clip= ImageClip(frame,ismask=False, transparent=True, fromalpha=False, duration=time)
  clip.write_videofile(nome+'.mp4',fps=1)
  videoclip = VideoFileClip(nome+'.mp4')
  audioclip = AudioFileClip(audio)
  new_audioclip = CompositeAudioClip([audioclip])
  videoclip.audio = new_audioclip
  videoclip.write_videofile(nome+'_'+'.mp4')
  lista_concat.append(nome+'_'+'.mp4')
  lista_mp3_mp4.append(nome+'_'+'.mp4')
  lista_mp4.append(nome+'.mp4')
  os.remove(audio)
  os.remove(frame)
  os.remove(nome+'.mp4')

def fim_clip (lsita_videos):
  movie=[]
  print(lsita_videos)
  x=0
  while(x<len(lsita_videos)):
    video = VideoFileClip(lsita_videos[x])
    movie.append(video)
    x=x+1
  final = concatenate_videoclips(movie, method='compose')
  final.write_videofile('final.mp4') 
```
